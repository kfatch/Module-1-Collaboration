```python
secret = 7
guess = 3
if guess < secret:
    print("Too low.")
elif guess > secret:
    print("Too high.")
elif guess == secret:
    print("Just right.")
else:
    print("Enter a valid guess.")
```

```python
number = 1
guess_me = 7
while number > 0:
    if number < guess_me:
        print("Too low.")
    elif number == guess_me:
        print("Found it!")
        break
    elif number > guess_me:
        print("Oops")
        break
    number += 1
```

```python
guess_me = 5
for num in range(10):
    if num < guess_me:
        print("Too low.")
    elif num == guess_me:
        print("Found it!")
        break
    elif num > guess_me:
        print("Oops.")
        break
```
